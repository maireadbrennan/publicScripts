```python
try:
    # Read and write names of input and output shapefiles
    # nameOfInputFeatureLayer     = os.path.basename(Polygon_Plot)
    centroidDistance            = float(1/2)  
    SmallCentroids              = "SmallCentroids"
    Polygon_Shrunk              = "Polygon_Shrunk"
    print('\n' + "Input shapefile: \t" + nameOfInputFeatureLayer)
    print('\n' + "Enlargement Factor: \t" + str(centroidDistance))
    print("Output shapefile: \t" + SmallCentroids + "\n")

    # Initialize an object to temporarily hold each new point as it is created
    newPoint1                   = arcpy.Point()

    # Initialize a list to hold all of the new features to be created
    listOfNewFeatures           = []
    listOfPoints                = []
    
    # Get the input shapefile's attribute table and the name of its shape field
    nameOfShapeField            = "SHAPE@"
    #arcpy.Describe(nameOfInputFeatureLayer).shapeFieldName
    print(nameOfShapeField)
    
    with arcpy.da.SearchCursor(nameOfInputFeatureLayer, ["OID@" , "SHAPE@", "SHAPE@AREA"]) as cursor:
        # Loop through the records of the input shapefile's attribute table, i.e. its features
        for row in cursor:
            # Get the next record's feature and its centroid point
            print('Feature {} has an area of {}'.format(row[0], row[2]))
            print("FEATURE (FID = " + str(row[0]) + ")")
            nextFeature             = row[1]
            print(str(nextFeature))
            pointAtCenter           = row[1].centroid
            print(str(pointAtCenter))

            # Initialize an array to hold the islands (i.e. parts) of a new feature to be created
            arrayOfNewIslands       = arcpy.Array()

            # Cycle through the islands of the current feature
            for nextIsland in nextFeature:
                print("\tISLAND:")

                # Initialize an array to hold the points for a new island to be created
                arrayOfNewPoints    = arcpy.Array()

                # Cycle through original island's vertices, creating a new point from each
                for nextVertex in nextIsland:
                    if nextVertex.X < pointAtCenter.X and nextVertex.Y > pointAtCenter.Y:
                        print("\t\t(" + str(nextVertex.X) + ',' + str(nextVertex.Y) + ")")
                        # If the next vertex is non-Null, create a new point and add it to the array of new points
                        newPoint1.X  = nextVertex.X + centroidDistance
                        newPoint1.Y  = nextVertex.Y - centroidDistance
                        arrayOfNewPoints.add(newPoint1)
                    elif nextVertex.X < pointAtCenter.X and nextVertex.Y < pointAtCenter.Y:
                        print("\t\t(" + str(nextVertex.X) + ',' + str(nextVertex.Y) + ")")
                        # If the next vertex is non-Null, create a new point and add it to the array of new points
                        newPoint1.X  = nextVertex.X + centroidDistance
                        newPoint1.Y  = nextVertex.Y + centroidDistance
                        arrayOfNewPoints.add(newPoint1)
                    elif nextVertex.X > pointAtCenter.X and nextVertex.Y > pointAtCenter.Y:
                        print("\t\t(" + str(nextVertex.X) + ',' + str(nextVertex.Y) + ")")
                        # If the next vertex is non-Null, create a new point and add it to the array of new points
                        newPoint1.X  = nextVertex.X - centroidDistance
                        newPoint1.Y  = nextVertex.Y - centroidDistance
                        arrayOfNewPoints.add(newPoint1)
                    elif nextVertex.X > pointAtCenter.X and nextVertex.Y < pointAtCenter.Y:
                        print("\t\t(" + str(nextVertex.X) + ',' + str(nextVertex.Y) + ")")
                        # If the next vertex is non-Null, create a new point and add it to the array of new points
                        newPoint1.X  = nextVertex.X - centroidDistance
                        newPoint1.Y  = nextVertex.Y + centroidDistance
                        arrayOfNewPoints.add(newPoint1)
                    else:
                        arcpy.AddMessage("\t\tHOLE: (beginning with a Null point)")
                        # If the next vertex is Null, insert a new point that is also Null
                        arrayOfNewPoints.append(None)   
                        
                # After creating an array of new points for a given island, add it to this feature's array of new islands 
                arrayOfNewIslands.append(arrayOfNewPoints)

            # After creating an array new islands for a given feature, create a new feature from that array
            newFeature = arcpy.Polygon(arrayOfNewIslands)

            # After creating a new feature, append it to a list of all new features
            listOfNewFeatures.append(newFeature)

        # Create the new shapefile from that list of all new features
        arcpy.CopyFeatures_management(listOfNewFeatures, Polygon_Shrunk)
        
        #Get Vertices from Polygon
        arcpy.AddField_management(Polygon_Shrunk, "SquareID", "FLOAT")
        pointer2 = 0
        with arcpy.da.UpdateCursor(Polygon_Shrunk, "SquareID") as cursors:
            for rows in cursors:
                rows[0] = SqID_List[pointer2]
                pointer2 += 1
                cursors.updateRow(rows)
        
            del rows
        del cursors
        
        # arcpy.FeatureVerticesToPoints_management(Polygon_Shrunk, SmallCentroids, "ALL")
        arcpy.FeatureVerticesToPoints_management(Polygon_Shrunk, SmallCentroids, "ALL")
        
        # Delete row and update cursor objects to avoid locking attribute table
    del nextIsland
    del nextVertex
    del row
    del cursor

except Exception as e:
    # If unsuccessful, end gracefully by indicating why
    arcpy.AddError('\n' + "Script failed because: \t\t" + str(e) )
    # ... and where
    exceptionreport = sys.exc_info()[2]
    fullermessage   = traceback.format_tb(exceptionreport)[0]
    arcpy.AddError("at this location: \n\n" + fullermessage + "\n")
```
